## Outliers

```python
#find outliers with IQR for price
Q1 = cars_db['Price($)'].quantile(0.25)
Q3 = cars_db['Price($)'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

price_outliers = cars_db[(cars_db['Price($)'] < lower) | (cars_db['Price($)'] > upper)]
price_outliers.count()
```

```python
sns.boxplot(cars_db['Price($)'])
```

```python
#find outliers with IQR for income
Q1 = cars_db['Annual_Income'].quantile(0.25)
Q3 = cars_db['Annual_Income'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

income_outliers = cars_db[(cars_db['Annual_Income'] < lower) | (cars_db['Annual_Income'] > upper)]
income_outliers.count()
```

```python
plt.boxplot(cars_db['Annual_Income'])
```

## Customer Analysis

>>What is the average annual income of customers?

```python
average_annual_income = cars_db['Annual_Income'].mean()
print(f"The average annual income of customers is: ${average_annual_income:.2f}")
```

```python
avg_income_by_gender = cars_db.groupby('Gender')['Annual_Income'].mean()
for gender, avg_income in avg_income_by_gender.items():
    print(f"The average annual income for {gender} customers is: ${avg_income:.2f}")
```

```python
avg_income_by_gender.plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.xlabel('Gender')
plt.ylabel('Average Annual Income ($)')
plt.title('Average Annual Income by Gender')
```

There is a notable income gap between male and female customers, with males earning more. Tailored marketing and product strategies can address their different financial capacities and preferences for better engagement.

>>What is the distribution of customers by gender?

```python
sales_by_gender = cars_db['Gender'].value_counts()
plt.bar(sales_by_gender.index, sales_by_gender.values, color=['pink', 'lightblue'])
plt.title('Sales Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Sales')
```

>>How many customers have an annual income above $1,000,000?

31.6% of customers earn over $1,000,000 annually, highlighting a wealthy segment. This creates opportunities to offer premium products for high earners while also catering to lower-income customers with affordable options.

>>Which dealer has the highest number of sales?

```python
dealer_sales = cars_db.groupby('Dealer_Name').size().reset_index(name='Sales')
highest_sales_dealer = dealer_sales.loc[dealer_sales['Sales'].idxmax()]
highest_sales_dealer
```

```python
ax = sns.barplot(x='Dealer_Name', y='Sales', data=dealer_sales, hue='Dealer_Name', palette='viridis')
plt.title('Number of Sales by Dealer')
plt.xlabel('Dealer Name')
plt.ylabel('Number of Sales')
plt.xticks(rotation=90)

# Highlight the bar with the highest sales
for bar in ax.patches:
    if bar.get_height() == highest_sales_dealer['Sales']:
        bar.set_color('red')
```

```python
print(f"The dealer with the highest number of sales is {highest_sales_dealer['Dealer_Name']} with {highest_sales_dealer['Sales']} sales.")
```

>>Do some dealers specialize in selling specific brands/models?

```python
dealer_brand_sales = cars_db.groupby(['Dealer_Name', 'Company', 'Model']).size().reset_index(name='Sales')
specialized_dealers = dealer_brand_sales[dealer_brand_sales['Sales'] > dealer_brand_sales['Sales'].mean()]
specialized_dealers.head()
```

```python
pivot_data = specialized_dealers.pivot_table(
    index='Dealer_Name', # the rows 
    columns=['Company', 'Model'], 
    values='Sales',             # Number of Sales
    fill_value=0                # Fill missing values with 0
)
```

```python
pivot_data.plot(kind='bar', stacked=True, figsize=(14, 8), colormap='viridis')
plt.title('Sales by Dealer, Brand, and Model (Specialized Dealers)')
plt.xlabel('Dealer Name')
plt.ylabel('Number of Sales')
plt.legend(bbox_to_anchor=(0.5, -0.2), loc='upper center', ncol=4)
```

The analysis of dealer sales data indicates that some dealers specialize in selling specific brands and models. For instance, Buddy Snobeck's Diesel Service Inc. focuses heavily on brands like BMW and Audi, with notable sales for models such as the BMW 3218 and Audi 86. This specialization likely stems from factors like expertise, customer trust, or targeted marketing.

Moreover, while some dealers maintain a diverse portfolio, others concentrate on a narrower range, potentially influenced by regional demand, dealer agreements, or customer preferences. This insight can help optimize sales strategies and resource allocation across different dealers.

## Date and Time Analysis

>>What is the busiest month for car sales?

```python
monthly_sales = cars_db.groupby('Month_ID').size()
monthly_sales.plot(kind='bar')
plt.title('Car Sales by Month')
plt.xlabel('Month')
plt.ylabel('Number of Sales')
x = np.arange(len(monthly_sales))  # Numeric representation of Month_ID
y = monthly_sales.values           # Sales data
m, b = np.polyfit(x, y, 1)         # Slope (m) and intercept (b)
plt.plot(x, m*x + b, color='orange', linestyle='--', label='Trendline')
```

>>Are there any trends in sales over time (e.g., increasing, decreasing)?

```python
daily_sales = cars_db.groupby('Date').size()
daily_sales.plot()
plt.title('Car Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Sales')
```

>>Which day of the week has the highest number of sales?

```python
cars_db['DayOfWeek'] = cars_db['Date'].dt.day_name()
daily_sales = cars_db.groupby('DayOfWeek').size()
daily_sales.plot(kind='bar')
plt.title('Car Sales by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Sales')
```

>>How many cars were sold on the busiest day?

```python
busiest_day = daily_sales.idxmax()
busiest_day_sales = daily_sales.max()
print(f"The busiest day for car sales is {busiest_day} with {busiest_day_sales} cars sold.")
```

>>How does the average car price vary across different months of the year?

```python
avg_price_by_month = cars_db.groupby('Month_ID')['Price($)'].mean().reset_index()
avg_price_by_month['Month_ID'] = avg_price_by_month['Month_ID']
plt.plot(avg_price_by_month['Month_ID'], avg_price_by_month['Price($)'], marker='o')
plt.title('Average Car Price Over Time')
plt.xlabel('Month')
plt.ylabel('Average Car Price ($)')
```

Monthly sales data shows fluctuations, with an overall trend indicating whether sales are increasing or decreasing. Daily sales trends highlight specific peak periods, with **Tuesday** emerging as the busiest day, recording **4,435 cars sold**. Sales by day of the week further confirm that weekdays, especially Tuesday, drive higher sales compared to weekends. Additionally, the average car price remains relatively stable across months, ranging between**27,200 and 28,600**, suggesting consistent pricing strategies.
