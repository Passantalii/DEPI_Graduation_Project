## Importing Libraries

```python
import pandas as pd 
import numpy as np
```

## Load Data

```python
file = 'Cleaned_data.csv'
cars_db = pd.read_csv(file)
```

```python
cars_db.head()
```

```python
cars_db['Dealer_Region'].unique()
```

## Dictionary of mapping

```python
city_to_state = {
    "Aurora": "Colorado",
    "Austin": "Texas",
    "Greenville": "South Carolina",
    "Janesville": "Wisconsin",
    "Middletown": "New York",
    "Pasco": "Washington",
    "Scottsdale": "Arizona",
}
```

```python
cars_db['Dealer_State'] = cars_db['Dealer_Region'].map(city_to_state)
```

```python
cars_db.head(3)
```

## Correct New Column Place

```python
ordered_cars_db = ['Customer_Name', 'Phone', 'Gender', 'Annual_Income', 'Car_id', 'Model', 'Color', 'Engine', 'Body_Style', 'Company', 'Price($)', 'Transmission', 'Dealer_No', 'Dealer_Name', 'Dealer_Region', 'Dealer_State', 'Date', 'Year_ID', 'Quarter_ID', 'Month_ID', 'Day_ID']
cars_db = cars_db[ordered_cars_db]
```

```python
cars_db.head(3)
```

## Save final Result

```python
cars_db.to_csv("Updated_Cleaned_data.csv", index=False)
```
